```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data/nyc_weather.csv',parse_dates=['EST'])
df.head()
```

```python
plt.figure(figsize=(20,10)) 
plt.xlabel('Day')
plt.ylabel('Value')
plt.title('Weather')
plt.plot(df['EST'],df['Temperature'], color='green', marker='*',linestyle='-.',markersize=10,label='Temperature')
plt.plot(df['EST'],df['DewPoint'], color='red', marker='D',linestyle='-.',markersize=10,label='DewPoint')
plt.legend(fontsize='large')
plt.grid()
```

```python
company_results = [
    {'company':'GOOGL','revenue':90,'profit':40},
    {'company':'AMZN','revenue':136,'profit':2},
    {'company':'MSFT','revenue':89,'profit':34},
    {'company':'FB','revenue':27,'profit':12},
]

company_df = pd.DataFrame(company_results)
company_df.head()
```

```python
xpos = np.arange(company_df.shape[0])
plt.bar(xpos,company_df['revenue'])
```

```python
xpos = np.arange(company_df.shape[0])
plt.xticks(xpos,company_df['company'])
plt.ylabel('revenue(bln)')
plt.title('US Stock Chart')
plt.bar(xpos,company_df['revenue'])
```

```python
xpos = np.arange(company_df.shape[0])
plt.xticks(xpos,company_df['company'])
plt.title('US Stock Chart')
plt.bar(xpos-0.2,company_df['revenue'],width=0.4,label='Revenue')
plt.bar(xpos+0.2,company_df['profit'],width=0.4,label='Profit')
plt.legend()
```

```python
xpos = np.arange(company_df.shape[0])
plt.yticks(xpos,company_df['company'])
plt.title('US Stock Chart')
plt.barh(xpos-0.2,company_df['revenue'],height=0.4,label='Revenue')
plt.barh(xpos+0.2,company_df['profit'],height=0.4,label='Profit')
plt.legend()
```
